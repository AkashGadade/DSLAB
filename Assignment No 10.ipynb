```python
import seaborn as sns
import matplotlib.pyplot as plt                                                                                                                                                                                                                                                                                                                                                                                                                       
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df.dtypes
```

```python
sns.histplot(df['sepal_length'],kde=True)
plt.title('distributionof sepal length')
plt.show()

sns.histplot(df['petal_length'],kde=True)
plt.title('distributionof petal length')
plt.show()

sns.histplot(df['sepal_width'],kde=True)
plt.title('distributionof sepal_width')
plt.show()

sns.histplot(df['petal_width'],kde=True)
plt.title('distributionof petal_width')
plt.show()
```

```python
sns.boxplot(df['sepal_length'])
plt.title('distributionof sepal length')
plt.show()

sns.boxplot(df['petal_length'])
plt.title('distributionof petal length')
plt.show()

sns.boxplot(df['sepal_width'])
plt.title('distributionof sepal_width')
plt.show()

sns.boxplot(df['petal_width'])
plt.title('distributionof petal_width')
plt.show()
```

```python
sns.pairplot(data=df,vars=['sepal_length','sepal_width'] ,hue='species')
```
