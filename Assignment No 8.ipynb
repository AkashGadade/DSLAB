```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Titanic.csv')
```

```python
df
```

```python
df2 = sns.load_dataset('titanic')
```

```python
df2
```

```python
df2.describe()
```

```python
# Paiplot to visualize relation between different numeric variables

sns.pairplot(df2, vars=['age','fare','survived'],hue='survived')
plt.title('Pairplot for titanic dataset')
plt.show()
```

```python
# heatmap to visualize correlation between variables

correlation = df2.corr()
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title('correalation heatmap')
plt.show()
```

```python
# countplot to visualize categorical variables

sns.countplot(x='class',hue='survived',data=df2)
plt.show()
```

```python
#violinplot to visualize distribution of age by survival

sns.violinplot(x='sex',y='age',data=df2)
plt.show()
```

```python
#boxplot to visualize the outliers

sns.boxplot(x='sex',y='age',data=df2)
```

```python
# stripplot

sns.stripplot(x='sex',y='age',data=df2,hue='survived')
```

```python
# swarmplot

sns.swarmplot(x='sex',y='age',data=df2,hue='survived')
```

```python
# catplot 
sns.catplot(x='sex',y='age',data=df2,hue='survived',kind='box')
```

```python
sns.catplot(x='sex',y='age',data=df2,hue='survived')
```

```python
# prise distribution by passengers

plt.figure(figsize=(16,5))
sns.histplot(x='fare',data=df2,kde=True)
```
